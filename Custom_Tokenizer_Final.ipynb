```python
import sentencepiece as spm
import os
```

```python
model_path = "/content/drive/MyDrive/Datasets/Datasets/Malayalam_Large_Corpus/malayalam_spm_large.model"

# Some interesting Malayalam sentences to test the tokenizer with.
# We've included a complex word from your corpus.
sample_texts = [
    "അവൻ പുസ്തകം വായിച്ചു.",
    "അതിവേഗം പരിണാമങ്ങളെ സ്വാംശീകരിക്കാനും",
    "ഉടലിനെ ജലം വായിച്ചതെങ്ങനെ എന്ന് ഉൽക്കണ്ഠയോടെ ഉറ്റുനോക്കുകയാണ്.",
    "2025-ൽ കേരളത്തിൽ എന്തു സംഭവിക്കും?"
]

# --- Main Test Logic ---

# Check if the model file exists before trying to load it.
if not os.path.exists(model_path):
    print(f"Error: Model file not found at '{model_path}'")
    print("Please make sure the file name and path are correct.")
else:
    print(f"✅ Found tokenizer model: '{model_path}'")

    # 1. Load the trained SentencePiece model.
    tokenizer = spm.SentencePieceProcessor()
    tokenizer.load(model_path)

    print(f"\nVocabulary size: {tokenizer.vocab_size()}")
    print("-" * 50)

    # 2. Loop through each sample sentence and test encoding/decoding.
    for text in sample_texts:
        print(f"Original Sentence: \n'{text}'\n")

        # 2a. Encode the text into subword "pieces".
        # This is great for human inspection.
        encoded_pieces = tokenizer.encode_as_pieces(text)
        print(f"Encoded as Pieces: \n{encoded_pieces}\n")

        # 2b. Encode the text into numerical IDs.
        # This is the actual input your TrOCR model will receive.
        encoded_ids = tokenizer.encode_as_ids(text)
        print(f"Encoded as IDs: \n{encoded_ids}\n")

        # 3. Decode the numerical IDs back to a text sentence.
        decoded_text = tokenizer.decode(encoded_ids)
        print(f"Decoded back to Text: \n'{decoded_text}'\n")

        # Verify that the decoded text matches the original.
        # assert text == decoded_text

        print("-" * 50)

    print("✅ All tests passed!")
```

```python
print("\n=== Tokenization Results ===")

for i, sentence in enumerate(malayalam_test_texts, 1):
    print(f"\n{i}. Original: {sentence}")

    # Encode to tokens (subwords)
    tokens = sp.encode(sentence, out_type=str)
    print(f"   Tokens: {tokens}")

    # Encode to IDs
    token_ids = sp.encode(sentence, out_type=int)
    print(f"   Token IDs: {token_ids}")
    print(f"   Token count: {len(token_ids)}")

    # Decode back to text
    decoded = sp.decode(token_ids)
    print(f"   Decoded: {decoded}")

    # Check if round-trip is perfect
    match = sentence == decoded
    print(f"   Round-trip match: ✅" if match else f"   Round-trip match: ❌")

    if not match:
        print(f"   ⚠️  Original: '{sentence}'")
        print(f"   ⚠️  Decoded:  '{decoded}'")

# Test special tokens
print(f"\n=== Special Tokens ===")
special_tokens = ["<PAD>", "<UNK>", "<BOS>", "<EOS>"]

for token in special_tokens:
    try:
        token_id = sp.piece_to_id(token)
        back_to_token = sp.id_to_piece(token_id)
        print(f"{token}: ID={token_id}, Back='{back_to_token}'")
    except:
        print(f"{token}: Not found in vocabulary")
```
